## Numpy Array and Basics

Learn About Numpy

```python
import numpy as np
```

### Creating array from list    

```python
arr_1d = np.array([1, 2, 3, 4, 5])
print("1d_array:" , arr_1d)

arr_2d = np.array([[1, 2, 3] , [4, 5, 6]])
print("2d_array:" , arr_2d)
```

### List vs array

```python
list_py = [1, 2, 3]
print("Python list multiplication", list_py * 2)

list_numpy = np.array([1, 2, 3])  
print("Numpy list multiplication", list_numpy * 2)

import time
start = time.time()
py_list = [i*2 for i in range(1000000)]
print("\n Python List operation time: ", time.time() - start)

start = time.time()
np.array = np.arange(1000000) * 2
print("\n Numpy List operation time: ", time.time() - start)


```

### Creating array from scratch

```python
Ones = np.ones((2,3))
print("Ones Array: \n", Ones)

Zeros = np.zeros((2,3))
print("Zeros Array: \n", Zeros)
```

### Vectors, Matrices and Tensor

```python
vector = np.array([1, 2, 3, 4, 5])  # 1D Array
print("Vector: ", vector)

Matrix = np.array([[1,2], [3,4]])  # 2D Arrays
print("Matrix: ", Matrix)

Tensor = np.array([[[1,2], [3,4]],[[5,6], [6,7]]]) # Combination of vector and Matrices
print("Tensor: ", Tensor)
```

### Array properties

```python
arr = np.array([[1, 2, 3], 
                [4, 5, 6]])

print("Shape" , arr.shape)
print("Size" , arr.size)
print("Dimesion" , arr.ndim)
print("Data Type" , arr.dtype)
```

### Array Resphaping

```python
arr = np.arange(12)
print(arr)

Reshaped = arr.reshape(3, 4)
print("Reshaped Array: \n", Reshaped)

Transpose = Reshaped.T
print("Transposed Array: \n", Transpose)

flattened = Reshaped.flatten()
print("Flattened Array: \n", flattened)
```
