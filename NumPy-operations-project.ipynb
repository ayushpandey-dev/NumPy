#### Numpy operations - Project

```python
import numpy as np 
import matplotlib.pyplot as plt
```

```python
# Data Structure: [restaurant_id, 2021, 2022, 2023, 2024]
sales_data = np.array([
    [1, 150000, 180000, 220000, 250000], #McDonalds
    [2, 120000, 140000, 160000, 190000], #BurgerKing
    [3, 200000, 230000, 260000, 300000], #Pizza Hut
    [4, 180000, 210000, 240000, 270000], #Dominos
    [5, 160000, 185000, 205000, 230000]  #Starbucks
    ])

print("==== Zomato Sales Analysis ====")
print("\n Sales Data Shape: ", sales_data.shape)
print("\n Sales of first 3 Restaurants: ", sales_data[:3])
```

```python
# total Yearly sales
Yearly_total = np.sum(sales_data[:, 1:], axis = 0) # axis=0 calculates values in row
print(Yearly_total)

Yearly_total = np.sum(sales_data[:, 1:], axis = 1) # axis=1 calculates values in Column
print(Yearly_total)

# Minimum sales per restarant
min_sales = np.min(sales_data[:, 1:], axis=1)
print(min_sales)
```

```python
# cummulative sum of Sales
cumsum = np.cumsum(sales_data[:, 1:], axis=1)
print(cumsum)

plt.Figure(figsize=(4, 6))
plt.plot(np.mean(cumsum, axis=0))
plt.title("Zomato Yearly Cummulative Sales")
plt.xlabel("Years")
plt.ylabel("Yearly Sales")
plt.grid(True)
plt.show()

plt.Figure(figsize=(4, 6))
plt.plot(np.sum(sales_data, axis=1))
plt.title("Zomato Yearly Cummulative Sales")
plt.xlabel("Years")
plt.ylabel("Yearly Sales")
plt.grid(True)
plt.show()
```

### Vectors

```python
vector1 = np.array([1, 2, 3, 4, 5])  
vector2 = np.array([6, 7, 8, 9, 10])

print("Vector Addition", vector1 + vector2)
print("Vector Multiplication", vector1 * vector2) 
```

```python
restaurant_type = np.array(['biryani', 'maggie', 'cake', 'pizza', 'bread'])

vectorized = np.vectorize(str.upper) # after using vectorized we can use any python basic operations on vector, str.uppar used to make the array data in uppercase

print("vectorized_upper", vectorized(restaurant_type))
```
